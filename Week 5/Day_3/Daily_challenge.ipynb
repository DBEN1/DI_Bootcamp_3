```python
# 1. Import the numpy, pandas, and matplotlib.pyplot packages.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 2. Use numpy to create an array x of 100 numbers ranging from 0 to 20 (inclusive) using the linspace function.
x = np.linspace(0, 20, 100)

# 3. Create an array y which is the square of x.
y = x**2

# 4. Use matplotlib to plot x vs y.
plt.plot(x, y, label="Plot of x vs y (y=x^2)", color='blue')

# 5. Your plot should have an appropriate title and labels for the x and y axes. 
# You should also adjust the x and y limits to correctly show all the data.
plt.title("y = x^2 Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 20)
plt.ylim(0, y.max())

# 6. Save this plot as a .png file using savefig.
plt.savefig("x_vs_y_squared.png")

# Display the plot
plt.show()
```
