```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

# Create a line plot
plt.plot(x, y)

# Set labels for x and y axis
plt.xlabel('Prices')
plt.ylabel('Demand')

# Set title for the plot
plt.title("Product Demand vs Price")
plt.savefig('product_demand_vs_price.png')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

# Create a scatter plot
plt.scatter(x, y)

# Set labels for x and y axis
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')

# Set title for the plot
plt.title('Children’s Height vs Age')
plt.savefig('children_height.png')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 2 * np.pi, 100)  # 100 evenly-spaced values from 0 to 2π

# Create a line plot for the sine function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, np.sin(x), color='tab:blue')
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Create a scatter plot for the cosine function
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.scatter(x, np.cos(x), color='tab:red', marker='o')
plt.title("Cosine Function")
plt.xlabel("x")
plt.ylabel("cos(x)")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 2 * np.pi, 100)  # 100 evenly-spaced values from 0 to 2π

# Create a figure with two subplots
fig, axs = plt.subplots(2)

# Create a line plot for the sine function in the first subplot
axs[0].plot(x, np.sin(x), color='tab:blue', label='y = sin(x)')
axs[0].set_title("Sine Function")
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")
axs[0].legend()  # show legend


# Create a scatter plot for the cosine function in the second subplot
axs[1].scatter(x, np.cos(x), color='tab:red', marker='o', label='y = cos(x)')
axs[1].set_title("Cosine Function")
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")
axs[1].legend()  # show legend


# Display the plots
plt.tight_layout()
plt.show()
```
